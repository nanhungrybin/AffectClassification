```python
import pandas as pd

# 파일 경로
file_path = '/workspace/ExpW/label/label.lst' 

# 파일 읽기
with open(file_path, 'r') as file:
    lines = file.readlines()


columns = ['image', 'IDlabel', 'x1', 'y1', 'x2', 'y2', 'confidence', 'explabel']
data = [line.split() for line in lines]
df = pd.DataFrame(data, columns=columns)

# 필요한 경우 숫자 열을 숫자로 변환
numeric_columns = ['IDlabel', 'x1', 'y1', 'x2', 'y2', 'confidence', 'explabel']
df[numeric_columns] = df[numeric_columns].apply(pd.to_numeric)

df
```

```python
"""
for expression label
"0" "angry"
"1" "disgust"
"2" "fear"
"3" "happy"
"4" "sad"
"5" "surprise"
"6" "neutral"

"""
import os
import pandas as pd

ExpW = pd.DataFrame(columns=['path', 'label'])

data_path = "/workspace/ExpW/origin"

# 이미지 이름이 df["image"]에 있는 경우
matching_images = df[ df["image"].isin(os.listdir(data_path)) ]

for _, row in matching_images.iterrows():
    img_path = os.path.join(data_path, row["image"])

    # 경로 및 레이블 정보를 추가
    ExpW = ExpW.append({'path': img_path, 'label': row["explabel"]}, ignore_index=True)

# 결과 확인
ExpW
```

```python
ExpW.to_csv("ExpW_annotation.csv")
```
