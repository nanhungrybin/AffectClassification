```python
import os

# "new_path" 열의 경로에서 역슬래시를 슬래시로 변경
df["path"] = df["path"].str.replace("\\", "/")

# 변경된 DataFrame 출력
print(df)
```

```python
# 기준 경로 설정
base_path_current = "D:/EXP_Data"
base_path_new = "/workspace"

# "new_path" 열 추가
df["new_path"] = ""

# 각 행에 대해 경로 변경 수행
for index, row in df.iterrows():
    current_path = row["path"]
    
    # 기존 경로에서 기준 경로를 삭제하고, 새로운 경로로 변경
    relative_path = os.path.relpath(current_path, base_path_current)
    new_path = os.path.join(base_path_new, relative_path)
    
    # 변경된 경로를 DataFrame에 저장
    df.at[index, "new_path"] = new_path
```

```python
df
```

```python
# "Unnamed: 0"와 "path" 열 제거
df = df.drop(columns=["Unnamed: 0", "path"])

# "new_path" 열을 DataFrame 앞쪽으로 이동
df = df[["new_path", "label"]]

```

```python

df = df.rename(columns={"new_path": "path"})

df
```
